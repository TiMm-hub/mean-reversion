```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('JenniferPriceData.csv')
data=data.iloc[:,1:]
```

```python
data.head(5)
```

## Start from data preprocessing

```python
# three matrix: return , price and sort, also the market capitalization matrix
```

```python
name,length=np.unique(data['Name'].values,return_counts=True)
max_length = max(length)
```

```python
%%time
price =[]
return_arr = []
mar_cap = []
for i in range(len(name)):
    df=data[data['Name']==name[i]]
    p=df['Price'].values
    r=(df['Price'].diff()/df['Price']).values
    cap=df['MarketCap'].values
    emp = np.empty((max_length-length[i],))
    emp[:] = np.NaN
    price.append(np.concatenate((emp,p)))
    return_arr.append(np.concatenate((emp,r)))
    mar_cap.append(np.concatenate((emp,cap)))
```

```python
replace_0_return=np.nan_to_num(return_arr,0)
sort=np.argsort(replace_0_return,axis=0)
```

```python
# market portfolio is useless, try to use the median return as baseline
```

```python
replace_0_cap=np.nan_to_num(mar_cap,0)
sort_cap=np.argsort(replace_0_cap,axis=0)
```

```python
median_arr=[]
for i in range(replace_0_return.shape[1]):
    for j in range(replace_0_return.shape[0]):
        if sort[j][i] == 326:
            median_arr.append(replace_0_return[j][i])
```

```python
median_arr
```

```python
np.cumsum(np.array(median_arr))
```

### Feature generation

```python
from scipy.ndimage.interpolation import shift
sort= -1*(sort-650)
sort_shift_1=shift(sort, [0,1], cval=0)/651
sort_shift_2=shift(sort, [0,2], cval=0)/651
sort_shift_3=shift(sort, [0,2], cval=0)/651
sort_shift_7=shift(sort, [0,7], cval=0)/651
```

```python
sort_diff = sort_shift_1-sort_shift_2
```

```python
sort_diff
```

```python
mar_cap
```

### define hyperparameter

```python
cap = 100 #  
t0 = 0 #  train start
t1 = 1200 # test start
t2 = np.shape(return_arr)[1]-1
lag = 1
bottom_num = 10
```

### select the bottom 10 with high marketcap and tagging

```python
def fetch_data_constraint(bottom_num,cap,lag,start,end):
    #sort_cap_rev= -1*(sort_cap-650)
    arr = []
    for j in range(start-lag,end):
        for i in range(np.shape(return_arr)[0]):
            if (sort_cap[i][j]<=cap) :
                if (sort[i][j]<bottom_num):
                    if not np.isnan(mar_cap[i][j]):
                        arr_ij=[sort_shift_1[i][j],sort_shift_2[i][j],sort_shift_3[i][j],sort_shift_7[i][j],sort_diff[i][j],mar_cap[i][j]]
                        arr_ij.append(1 if return_arr[i][j+lag]>=median_arr[j+lag] else 0)
                        arr.append(arr_ij)
    return arr
```

```python
train=fetch_data_constraint(bottom_num,cap,lag,t0,t1)
```

```python
np.unique(np.array(train)[:,6],return_counts=True)
```

```python
test=fetch_data_constraint(bottom_num,cap,lag,t1,t2)
```

```python
np.unique(np.array(test)[:,6],return_counts=True)
```

```python
x_train = np.array(train)[:,:5]
y_train = np.array(train)[:,6]
x_test = np.array(test)[:,:5]
y_test = np.array(test)[:,6]
```

## Conclusion:
To real implmentation, median return is not stable. As a result, I will use return=0.003 as baseline to cover the transaction cost.

```python
def fetch_data_constraint_new(bottom_num,cap,lag,start,end):
    #sort_cap_rev= -1*(sort_cap-650)
    arr = []
    arr_return = []
    for j in range(start-lag,end):
        for i in range(np.shape(return_arr)[0]):
            if (sort_cap[i][j]<cap) :
                if (sort[i][j]<bottom_num):
                    if not np.isnan(mar_cap[i][j]):
                        arr_ij=[sort_shift_1[i][j],sort_shift_2[i][j],sort_shift_3[i][j],sort_shift_7[i][j],sort_diff[i][j],mar_cap[i][j]]
                        arr_ij.append(1 if return_arr[i][j+lag]>=0.003 else 0)
                        arr_return.append(return_arr[i][j+lag])
                        arr.append(arr_ij)
    return arr,arr_return
```

```python
train,return_train=fetch_data_constraint_new(bottom_num,cap,lag,t0,t1)
test,return_test=fetch_data_constraint_new(bottom_num,cap,lag,t1,t2)
np.unique(np.array(train)[:,6],return_counts=True),np.unique(np.array(test)[:,6],return_counts=True)
```

```python
x_train = np.array(train)[:,:5]
y_train = np.array(train)[:,6]
x_test = np.array(test)[:,:5]
y_test = np.array(test)[:,6]
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(clf.predict(x_test),y_test)
```

```python
num_round = 100
bst = xgb.train(param, dtrain, num_round, evallist,early_stopping_rounds=30)
```
